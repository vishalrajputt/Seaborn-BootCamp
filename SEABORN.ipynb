```python
import seaborn as sns
import numpy as np
import pandas as pd
import scipy 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
titanic_ds = sns.load_dataset('titanic') 
```

```python
titanic_ds.head()
```

```python
sns.distplot(titanic_ds['fare'])
```

```python
sns.distplot(titanic_ds['fare'],kde=False,bins=50)
```

```python
sns.jointplot(x='age',y='fare',data=titanic_ds)
```

```python
sns.jointplot(x='sex',y='alive',data=titanic_ds)
```

```python
sns.jointplot(x='age',y='fare',data=titanic_ds,kind='hex')
```

```python
sns.jointplot(x='age',y='fare',data=titanic_ds,kind='reg')
```

**`Factor plot`** and **`Fat consumption data`**

```python
df = sns.load_dataset('exercise')
```

```python
df.head()
```

```python
df = sns.factorplot(x='time',y='pulse',hue='kind',data=df)
```

```python
df = sns.factorplot(x='time',y='pulse',hue='diet',data=df)
```

**`Swarmplot`** with IRIS dataset

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.swarmplot(x='species',y='petal_length',data=iris)
```

```python
sns.swarmplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.swarmplot(x='species',y='sepal_length',data=iris)
```
